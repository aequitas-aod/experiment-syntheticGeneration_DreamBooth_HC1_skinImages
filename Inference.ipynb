```python
import os
from google.colab import files
from PIL import Image
import glob
import locale
import torch
from diffusers import DiffusionPipeline, AutoencoderKL
```

# Inference with base model

```python
vae = AutoencoderKL.from_pretrained("madebyollin/sdxl-vae-fp16-fix", torch_dtype=torch.float16)
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    vae=vae,
    torch_dtype=torch.float16,
    variant="fp16",
    use_safetensors=True,
)
_ = pipe.to("cuda")
```

```python
prompt = "skin with disease papular maculo exanthema" # @param
generated_images = []
for i in range(100):
  image = pipe(prompt=prompt, num_inference_steps=100, height=256, width=256).images[0]
  generated_images.append(image)
```

# Inference with model fine-tuned with small dataset

```python
vae = AutoencoderKL.from_pretrained("madebyollin/sdxl-vae-fp16-fix", torch_dtype=torch.float16)
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    vae=vae,
    torch_dtype=torch.float16,
    variant="fp16",
    use_safetensors=True,
)
pipe.load_lora_weights("/gdrive/MyDrive/AIIIProject/normal_training")
_ = pipe.to("cuda")
```

```python
prompt = "papular maculo exanthema" # @param
generated_images = []
for i in range(100):
  image = pipe(prompt=prompt, num_inference_steps=100, height=256, width=256).images[0]
  generated_images.append(image)
```
